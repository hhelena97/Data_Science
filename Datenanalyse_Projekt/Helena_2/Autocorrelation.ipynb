# Autokorrelation und partielle Autokorrelation

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

#Trainingsdaten
df_training = pd.read_csv('training_data_integrated.csv', index_col='Datetime', parse_dates=True)
```

## Fahrradzählungen:
#### Tag:

```python
#Korrelation über den Tag (24 Stunden)
plot_acf(df_training['Bonn2016-2021'], lags=24)
plot_pacf(df_training['Bonn2016-2021'], lags=24)
plt.show()
```

#### Monat:

```python
#Korrelation über den Monat (ca 720 Std.)
plot_acf(df_training['Bonn2016-2021'], lags=720)
plot_pacf(df_training['Bonn2016-2021'], lags=720)
plt.show()
```

```python
#Täglich über den Monat hinweg
plot_acf(df_training['Bonn2016-2021'].resample('D').sum(), lags=30)
plot_pacf(df_training['Bonn2016-2021'].resample('D').sum(), lags=30)
plt.show()
```

#### Jahr:

```python
#Korrelation über das Jahr (8760 Std.)
plot_acf(df_training['Bonn2016-2021'], lags=8760)
plot_pacf(df_training['Bonn2016-2021'], lags=8760)
plt.show()
```

```python
#Tage übers Jahr
plot_acf(df_training['Bonn2016-2021'].resample('D').sum(), lags=365)
plot_pacf(df_training['Bonn2016-2021'].resample('D').sum(), lags=365)
plt.show()
```

```python
#Monate übers Jahr
plot_acf(df_training['Bonn2016-2021'].resample('M').sum(), lags=12)
plot_pacf(df_training['Bonn2016-2021'].resample('M').sum(), lags=12)
plt.show()
```

### Temperatur

```python
#über den Tag
plot_acf(df_training['Temperature (°C)'], lags=24)
plot_pacf(df_training['Temperature (°C)'], lags=24)
plt.show()
```

```python
#über den Monat
plot_acf(df_training['Temperature (°C)'].resample('D').sum(), lags=30)
plot_pacf(df_training['Temperature (°C)'].resample('D').sum(), lags=30)
plt.show()
```

```python
#übers Jahr
plot_acf(df_training['Temperature (°C)'].resample('M').sum(), lags=12)
plot_pacf(df_training['Temperature (°C)'].resample('M').sum(), lags=12)
plt.show()
```

#### Niederschlag:

```python
#übers Jahr
plot_acf(df_training['Precipitations (mm)'].resample('M').sum(), lags=12)
plot_pacf(df_training['Precipitations (mm)'].resample('M').sum(), lags=12)
plt.show()
```

#### Wind:

```python
plot_acf(df_training['Wind Speed (m/s)'].resample('M').sum(), lags=12)
plot_pacf(df_training['Wind Speed (m/s)'].resample('M').sum(), lags=12)
plt.show()
```
